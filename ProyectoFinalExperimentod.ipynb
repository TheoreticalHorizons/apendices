```python
# -*- coding: utf-8 -*-

!pip install openpyxl
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
g = 9.81


df1_B = pd.read_excel('DEFINITIVO_Datos_video1.xlsx', sheet_name='Masa A')
df1_A = pd.read_excel('DEFINITIVO_Datos_video1.xlsx', sheet_name='Masa B')
df2_B = pd.read_excel('DEFINITIVO_Datos_video2.xlsx', sheet_name='Masa A')
df2_A = pd.read_excel('DEFINITIVO_Datos_video2.xlsx', sheet_name='Masa B')

df1_A['angle'] = np.radians(df1_A['angle'])
df1_B['angle'] = np.radians(df1_B['angle'])
df1_A['w'] = np.radians(df1_A['w'])
df1_B['w'] = np.radians(df1_B['w'])
df2_A['angle'] = np.radians(df2_A['angle'])
df2_B['angle'] = np.radians(df2_B['angle'])
df2_A['w'] = np.radians(df2_A['w'])
df2_B['w'] = np.radians(df2_B['w'])

plt.figure(figsize=(12, 6))
plt.plot(df1_A['t'], df1_A['angle'], label='Varilla 1 (Superior)')
plt.plot(df1_B['t'], df1_B['angle'], label='Varilla 2 (Inferior)')
plt.title('Posicin Angular vs Tiempo - Condicin 1', fontsize=20)
plt.xlabel('Tiempo (s)', fontsize=20)
plt.ylabel('ngulo (radianes)', fontsize=20)
plt.legend()
plt.grid(True)
plt.legend(fontsize=14)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df2_A['t'], df2_A['angle'], label='Varilla 1 (Superior)')
plt.plot(df2_B['t'], df2_B['angle'], label='Varilla 2 (Inferior)')
plt.title('Posiciones Angulares vs Tiempo - Condicin 2', fontsize=20)
plt.xlabel('Tiempo (s)', fontsize=20)
plt.ylabel('ngulo (radianes)', fontsize=20)
plt.legend()
plt.grid(True)
plt.legend(fontsize=14)
plt.show()


max_t = min(df1_A['t'].max(), df2_A['t'].max())
plt.figure(figsize=(12, 6))
plt.plot(df1_A['t'], df1_A['angle'], label='Condicin 1 - Varilla 1')
plt.plot(df2_A['t'], df2_A['angle'], label='Condicin 2 - Varilla 1')
plt.xlim(0, max_t)
plt.title('Comparacin de Posicin Angular para la Varilla 1 (Se muestra Divergencia Catica)', fontsize=20)
plt.xlabel('Tiempo (s)', fontsize=20)
plt.ylabel('ngulo (radianes)', fontsize=20)
plt.legend()
plt.grid(True)
plt.legend(fontsize=14)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df1_B['t'], df1_B['angle'], label='Condicin 1 - Varilla 2')
plt.plot(df2_B['t'], df2_B['angle'], label='Condicin 2 - Varilla 2')
plt.xlim(0, max_t)
plt.title('Comparacin de Posicin Angular para la Varilla 2 (Se muestra Divergencia Catica)', fontsize=20)
plt.xlabel('Tiempo (s)', fontsize=20)
plt.ylabel('ngulo (radianes)', fontsize=20)
plt.legend()
plt.grid(True)
plt.legend(fontsize=14)
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(df1_A['angle'], df1_A['w'])
plt.title('Espacio Fsico para Varilla 1 - Condicin 1', fontsize=20)
plt.xlabel('ngulo (radianes)', fontsize=20)
plt.ylabel('Velocidad Angular (rad/s)', fontsize=20)
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(df1_B['angle'], df1_B['w'])
plt.title('Espacio Fsico para Varilla 2 - Condicin 1', fontsize=20)
plt.xlabel('ngulo (radianes)', fontsize=20)
plt.ylabel('Velocidad Angular (rad/s)', fontsize=20)
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(df2_A['angle'], df2_A['w'])
plt.title('Espacio Fsico para Varilla 1 - Condicin 2', fontsize=20)
plt.xlabel('ngulo (radianes)', fontsize=20)
plt.ylabel('Velocidad Angular (rad/s)', fontsize=20)
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(df2_B['angle'], df2_B['w'])
plt.title('Espacio Fsico para Varilla 2 - Condicin 2', fontsize=20)
plt.xlabel('ngulo (radianes)', fontsize=20)
plt.ylabel('Velocidad Angular (rad/s)', fontsize=20)
plt.grid(True)
plt.show()

m = 0.38
l = 0.31
g = 9.77


theta1 = df1_A['angle']
dtheta1 = df1_A['w']
theta2 = df1_B['angle']
dtheta2 = df1_B['w']
T = (2/3) * m * l**2 * dtheta1**2 + (1/6) * m * l**2 * dtheta2**2 + (1/2) * m * l**2 * np.cos(theta1 - theta2) * dtheta1 * dtheta2
V = -(3/2) * m * g * l * np.cos(theta1) - (1/2) * m * g * l * np.cos(theta2)
E_total1 = T + V
plt.figure(figsize=(12, 6))
plt.plot(df1_A['t'], E_total1)
plt.title('Energa Mecnica Total vs Tiempo - Condicin 1', fontsize=20)
plt.xlabel('Tiempo (s)', fontsize=20)
plt.ylabel('Energa (J)', fontsize=20)
plt.grid(True)
plt.show()


theta1 = df2_A['angle']
dtheta1 = df2_A['w']
theta2 = df2_B['angle']
dtheta2 = df2_B['w']
T = (2/3) * m * l**2 * dtheta1**2 + (1/6) * m * l**2 * dtheta2**2 + (1/2) * m * l**2 * np.cos(theta1 - theta2) * dtheta1 * dtheta2
V = -(3/2) * m * g * l * np.cos(theta1) - (1/2) * m * g * l * np.cos(theta2)
E_total2 = T + V
plt.figure(figsize=(12, 6))
plt.plot(df2_A['t'], E_total2)
plt.title('Energa Mecnica Total vs Tiempo - Condicin 2', fontsize=20)
plt.xlabel('Tiempo (s)', fontsize=20)
plt.ylabel('Energa (J)', fontsize=20)
plt.grid(True)
plt.legend(fontsize=14)
plt.show()
```
